```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
print(os.listdir("FB"))
```

```python
# IMPORTING DATA

data_train = pd.read_csv("FB/FB_Stock_Train.csv")
data_train
```

```python
trainset = data_train.iloc[:, 1:2].values
trainset
```

```python
#Normalize Data

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0,1))
trainData_scaled = sc.fit_transform(trainset)
```

```python
trainData_scaled
```

```python
x_train = []
y_train = []
```

```python
for i in range(60, 1414): #60 is the memory size, 1414 rows of training data
    x_train.append(trainData_scaled[i-60:i, 0]) # adds data for the last 60 rows
    y_train.append(trainData_scaled[i,0]) # adds the prices that the model is trying to predict

# Convert to tensors
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1)) # adds a 3rd dimension to the tensor
#                     array , new size -->(x, y, z)
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
model = Sequential()

model.add(LSTM(units = 50, return_sequences = True, input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 50))
model.add(Dropout(0.2))

model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, epochs = 100, batch_size = 32)
```

```python
model.save("FB_RNN.hd5f")
```

```python
data_test = pd.read_csv("FB/FB_Stock_Test.csv") # Import Test Data
```

```python
actual_stock_price = data_test.iloc[:, 1:2].values # Get the Open Prices
```

```python
data_total = pd.concat((data_train['Open'], data_test['Open']), axis=0) 
# combines the train and test data (axis=0 labels by index)
data_total[0:5]
```

```python
inputs = data_total[len(data_total) - len(data_test)-60:].values #removes the answers
print(len(data_total))
print(len(data_test))

inputs = inputs.reshape(-1, 1) #Converts the row of values into a column vector
inputs = sc.transform(inputs) # Feature Scaling (0-1)
inputs.shape
```

```python
# Prepare Testing Inputs
x_test = []

for i in range(60, inputs.shape[0]):
    x_test.append(inputs[i-60:i, 0]) #Prepares the 60 data memory input

x_test = np.array(x_test) #Turn the inputs into an array
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1)) # Add a 3rd dimension
```

```python
# Use the RNN to make the predeictions 
raw_prediction = model.predict(x_test)
```

```python
price_prediction = sc.inverse_transform(raw_prediction) # Undoes the normalization into usable values
price_prediction [0:5]
```

```python
sc2 = MinMaxScaler(feature_range = (-1,1))

raw_prediction = sc2.fit_transform(raw_prediction)

Suggestion = []
percentChange = []

for d in range(1,len(raw_prediction)):
    percentChange.append(float((raw_prediction[d]-raw_prediction[d-1])))
    
for n in percentChange:
    #print(n)
    if n > 0.015:
        Suggestion.append(-1) # Sell after a threshhold of increase
    elif n < 0.0:
        Suggestion.append(1) # Buy when there is a drop
    else:
        Suggestion.append(0)
#print(Buy)
```

```python
PurchaseHistory=[]
BuyPoints =  [[], []]
SellPoints = [[], []]
bought = False
for i in range(0, len(actual_stock_price)-1):
    if Suggestion[i] == 1 and not bought:
        PurchaseHistory.append(["Buy: ", actual_stock_price[i], i])
        BuyPoints[0].append(actual_stock_price[i])
        BuyPoints[1].append(i)
        bought = True
    if Suggestion[i] == -1 and bought:
        PurchaseHistory.append(["Sell:", actual_stock_price[i], i])
        SellPoints[0].append(actual_stock_price[i])
        SellPoints[1].append(i)        
        bought = False
if len(PurchaseHistory)%2 != 0:
    PurchaseHistory.pop()
    BuyPoints[0].pop()
    BuyPoints[1].pop() 
#PurchaseHistory
```

```python
Profit = 0
for purchase in range(1, len(PurchaseHistory)+1, 2):
    MoneyMade = float(PurchaseHistory[purchase][1] - float(PurchaseHistory[purchase-1][1]))
    Profit += MoneyMade

print('$'+ str(round(Profit, 2)))
```

```python
plt.plot(actual_stock_price, color =  'purple', label = "Actual Price")
plt.plot(price_prediction, color = 'blue', label = "Predicted Price")
plt.plot(BuyPoints[1], BuyPoints[0], "go")
plt.plot(SellPoints[1], SellPoints[0], "ro")

plt.title("Facebook January 2018 - February 2019 Stock Predictions")
plt.xlabel('Time')
plt.ylabel('Facebook Stock Price')
plt.legend()
plt.show()
```
